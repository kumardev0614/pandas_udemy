```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
pd.options.display.max_columns = 30
pd.options.display.float_format = '{:.2f}'.format
```

## Data Import

```python
df1 = pd.read_csv('movies_complete.csv', parse_dates=['release_date'])
```

```python
df1
```

```python
# we will create a new data frame with only columns we required
```

```python
df = df1[['title', 'belongs_to_collection', 'revenue_musd', 'vote_count', 'popularity']].copy()
```

```python
df
```

# Groupby: Normal Way

```python
g1 = df.groupby('belongs_to_collection').sum()
```

```python
g1

# Here We have grouped the data by belongs_to_collection.
# sum() will add the column values which belongs to same collection.
```

```python
g1.nlargest(10, 'popularity')
```

```python
g1.nlargest(10, 'revenue_musd')
```

# Groupby with multiple aggregate functions

```python
# It above method we can apply only one aggregate function. Also that aggregate function applies to all columns.
```

```python
# Now we will apply multiple aggrigate fuction on each column.
```

```python
g2 = df.groupby('belongs_to_collection').agg({"title":'count', "revenue_musd":'mean', "vote_count":'sum',
                                         "popularity":'mean'})
```

```python
g2
```

```python
g2.nlargest(10, 'revenue_musd')
```

### More than one aggregate function on a single column

```python
g3 = df.groupby('belongs_to_collection').agg({"title":'count', "revenue_musd":['mean', 'sum'], "vote_count":['sum', 'mean'],
                                         "popularity":'mean'})
```

```python
g3
```

```python
g3.nlargest(10, ('revenue_musd', 'mean'))   # Here we have to pass tuple of column name with outer index and inner index.
```

```python
g3.nlargest(10, ('revenue_musd', 'sum'))
```

# Third Way: Creating new columns to save aggregate result.

```python
g4 = df.groupby('belongs_to_collection').agg(No_of_movie=('title','count'),
                                     mean_revenue=('revenue_musd','mean'),
                                     total_revenue=('revenue_musd','sum'),
                                     arg_vote=('vote_count','mean'),
                                     total_popularity=('popularity','sum'))
```

```python
g4
```

```python
g4.nlargest(10, 'mean_revenue')  # Here we have removed multiple index.
```

```python
g4.nlargest(10, 'total_revenue')
```
