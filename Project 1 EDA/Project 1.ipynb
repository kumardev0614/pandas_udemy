### Make a copy of dataframe

```python
# First we will create a saperate copy of data then we will manupulate it.
# We will keep only those columns which are required for this analysis.
# bw - means best and worse

df_bw = df[['title', 'poster_path', 'budget_musd', 'revenue_musd', 'vote_count', 'vote_average', 'popularity' ]].copy()

df_bw
```

```python
# Now we will change the names of columns for our ease.

df_bw.columns = ['title', 'poster', 'budget', 'revenue', 'votes', 'rating', 'popularity']

df_bw.head(5)
```

```python
# set title as index of new dataframe

df_bw.set_index('title', inplace=True)

df_bw.head(5)
```

```python
# Now we will add 2 extra columns, profit and ROI.

df_bw['profit'] = df_bw.revenue.sub(df_bw.budget)
df_bw['ROI'] = df_bw.revenue.div(df_bw.budget)
```

```python
df_bw
```

## Highest Rating

## Highest ROI

```python
df_bw.sort_values(by='ROI', ascending=False)
```

```python
# 12396383.00 this much of ROI is not acceptable. Here the problem is, budget is 0.00 so we need to filter out here also.
# Now we will consider only those movies which have budget of at least 5 musd.

df_bw.loc[ df_bw.budget >= 5 ].sort_values(by='ROI', ascending=False)
```

```python
df_bw.votes.fillna(0, inplace=True)
df_bw.budget.fillna(0, inplace=True)
```

```python
# Now we will create a fuction which can find most of our queries.

def best_worst(n, by, ascending=False, min_b=0, min_v=0):
    df2 = df_bw.loc[ (df_bw.budget >= min_b) & (df_bw.votes >= min_v) ].sort_values(by=by, ascending=ascending).head(n).copy()
    return HTML(df2.to_html(escape=False))
```

```python
best_worst(5, 'ROI', min_b=1)
```

```python
best_worst(5, 'ROI', ascending=True)
```

```python
best_worst(5, "rating", min_v=300)
```

```python
best_worst(5, 'revenue')
```

```python
best_worst(5, 'revenue', True)
```

We can play with this fuction to see different results.

### Search 1: Science Fiction Action Movie with Bruce Willis (sorted from high to low Rating)

```python
# We have done this before in self coding project 1.
# Here we will use the method of course. Which is easy i think.

# First we will mask(filter out) a series of True/False which will further be used in whole dataframe for desired result.

mask_genres = df.genres.str.contains('Action') & df.genres.str.contains('Science Fiction')
mask_genres
```

```python
# Ok first condition is fullfilled now go for second condition.

mask_actor = df.cast.str.contains('Bruce Willis')
mask_actor
```

```python
# Now we will pass these 2 True/False series into df to get combind result.
# Here notice that we can pass the attributes in loc() function also.

df.loc[mask_genres & mask_actor, ['title', 'vote_average']].sort_values(by='vote_average', ascending=False)
```

```python
# Without loc[] function.

df[mask_genres & mask_actor][['title', 'vote_average']].sort_values(by='vote_average', ascending=False)
```

```python
# Without variables and loc.

df[ df.genres.str.contains('Action') & df.genres.str.contains('Science Fiction') & df.cast.str.contains('Bruce Willis') ][['title', 'vote_average']].sort_values(by='vote_average', ascending=False)
```

```python
# We can show images also by including poster_path column, then converting df into html, and then passing it to HTML() method.

bruce=df.loc[mask_genres & mask_actor, ['title', 'poster_path', 'vote_average']].sort_values(by='vote_average', ascending=False)
bruce
```

# Wordcloud

```python
from wordcloud import WordCloud
```

```python
df[['title', 'tagline', 'overview']]
```

```python
df.title[0]
```

```python
df.overview[0]
```

```python
# To make a cloud we don't need any NaN value. So we will drop all NAN values from these 3 columns and save them in as Series.
# because resultent will be a series.

titles = df.title.dropna()
tagline = df.tagline.dropna()
overview = df.overview.dropna()
```

```python
titles
```

```python
tagline              # We can see tagline has so many missing values.
```

```python
# Now we will create somthing which is called corpus. Mean a string containing all words saparated by " ". 
# This string will be passed to create the WordCloud from it.

title_corpus = " ".join(titles)
tagline_corpus = ' '.join(tagline)
overview_corpus = ' '.join(overview)
```

```python
title_corpus
```

### WordCloud

```python
title_wordcloud = WordCloud(background_color='white', height=2000, width=4000, max_words=500).generate(title_corpus)
title_wordcloud
```

```python
# Now we got wordcloud object. It is an image. We just have to plot it.

plt.figure(figsize=(16,8))
plt.imshow(title_wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
