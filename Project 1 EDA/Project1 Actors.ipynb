# Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
pd.options.display.max_columns = 30
pd.options.display.float_format = '{:.2f}'.format
```

### Import Data

```python
df = pd.read_csv('movies_complete.csv', parse_dates=['release_date'])
```

```python
df[['id', 'title', 'cast', 'revenue_musd', 'vote_average', 'popularity']]
```

### Resetting Index

```python
# First of all we change the index of our df to movie ID.
# Which we will use in below task.
```

```python
df.set_index('id', inplace=True)
```

```python
df[['title', 'cast', 'revenue_musd', 'vote_average', 'popularity']]
```

### Now we will create a new dataframe using names of all actors in df.cast. Then we will add the data in new dataframe corresponding to actors.

### Spliting each actor name in each row of cast column.

```python
# First we will split the cast column values and create a datafram from these splited values
```

```python
df.cast.str.split('|')
```

```python
# This a series but we need a DataFrame.
df.cast.str.split('|', expand=True)
```

```python
# So here we can see split method had created a column using each splited value. We got 313 columns because one of the movie
# in df.cast has 313 actor names.
# Here we got both None and NaN Both.
# None != NaN
# We got None because cast has only 10 to 12 actor names. That's why remaining columns are filled with 'None'
# We got NaN in all columns because df.cast is NaN. So we didn't got any value after NaN but NaN.
```

```python
# Now we will save the result in New Dataframe actors.
actors = df.cast.str.split('|', expand=True)
```

```python
# Ignore the NaN and None values. They will be removed in final dataframe. 
```

```python
actors
```

### Convert column values into rows.

```python
# Now we will change all these columns values into rows. So that we get a single column with all the actor names as rows.
actors.stack()
```

### Reset Multilevel Index

```python
# We got what we need. but stack() method has created inner indexes 0,1,2,3... which we dont need.
# So we will reset our index to outer index, and drop the inner index.
# Here we have 2 index levels. level 0(outer index) and level 1(inner index)
# here reset() method will take 2 arguments level(which has to be removed) and drop=True
```

```python
actors.stack().reset_index(level=1, drop=True)
```

### Convert Series into Data Frame

```python
# But we don't need a series we need a dataframe.

actors.stack().reset_index(level=1, drop=True).to_frame()
```

```python
actors.head()
```

```python
# We can see our actors dataframe is still same, because we havn't saved the changes.
# Here we can do all changes in sigle line of code.
# We have done it many lines just to show the changes.

actors = actors.stack().reset_index(level=1, drop=True).to_frame()
```

```python
actors
```

```python
# We will change the column name also
actors.columns = ['actors']

# it will change the 0th positoned column to 'actors'
```

```python
# Now our dataframe is ready to merge.
# We can see stack() method has droped all None and NaN values by itself.
# We have only 12 to 20 values for a single id.
actors.head(30)
```

## Merging, adding corresponding data into actors dataframe.

```python
# We will perform left join.

actors= actors.merge(df[['title', 'revenue_musd', 'vote_average', 'popularity']], how='left', left_index=True, right_index=True)
```

```python
actors
```

### Creating new Dataframe with more accurate data for individual actor.

```python
# Now we will groupby the data by actors so that we can analyse all the work of individual actor.
# We will use labeled aggrigated functions to avoid the 2 level column index.
actors_all = actors.groupby('actors').agg(total_revenue=('revenue_musd','sum'),
                                          mean_revenue =('revenue_musd','mean'),
                                          mean_rating = ('vote_average','mean'),
                                          mean_pop= ('popularity','mean'),
                                         total_movies = ('actors','count'))

# Here we are calculating mean with respect to each actor.
# So this new dataframe will be more accurate to show data of each actor.
```

```python
actors_all

# It's looking like a 2 level index but it's not.
```

### top 10 actors with most movies done.

```python
actors_all.nlargest(10, 'total_movies')
```
