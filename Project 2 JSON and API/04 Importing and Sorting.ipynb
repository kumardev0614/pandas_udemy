# Creating Query and getting the Response.

```python
# Now we will ask for movies data for 18 movies by providing there id.
# There also some wrong ids here.
```

```python
api_key = 'api_key=6b5924df7b618eb4d70fc47936cad3cf'

movie_id = [0, 299534, 19995, 140607, 299536, 597, 135397,
            420818, 24428, 168259, 99861, 284054, 12445,
            181808, 330457, 351286, 109445, 321612, 260513, 228965, 302589]
```

```python
query = 'https://api.themoviedb.org/3/movie/{}?{}'
```

### Code to save multiple response into single list.

```python
json_list = []
invalid_id = []
response_l = []

for m_id in movie_id:
    url = query.format(m_id, api_key)
    response = requests.get(url)
    
    response_l.append(response.status_code)
    
    if response.status_code != 200:
        invalid_id.append(m_id)
    else:
        data = response.json()
        json_list.append(data)       
```

```python
json_list 

# We got a dict in each response.
# So we have created a list of dicts
# json_list[ {id 299534 data}, {id 19995 data}, {id 140607 data}, {}, ..... ]
```
